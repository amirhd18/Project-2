```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#Load date in from CSV file

df_2019 = pd.read_csv("../Project_2/2019.csv")
df_2020 = pd.read_csv("../Project_2/2020.csv")
df_2021 = pd.read_csv("../Project_2/2021.csv")
```

```python
#show data is reading correctly for each file

df_2019.head()
```

```python
#combined all data into one data set

df = pd.concat([df_2019,df_2020,df_2021])
```

```python
#Add date time format

df['Date'] = pd.to_datetime(df['Date'])
```

```python
# break down dates to months and year to be able to group the months.

df['Month_Number'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df.head()
```

```python
df2 = df.groupby(['Year','Month_Number']).sum()
```

```python
df2.head()
```

```python
df2.reset_index(inplace = True)
```

```python
df2.columns
```

```python
df2.head()
```

```python
df2['ds']=df2['Month_Number'].map(str) + '/' + df2['Year'].map(str)
```

```python
df2.head()
```

```python
df2.drop(['Year','Month_Number'], axis =1)
```

```python
df2=df2[['ds','Amount']]
```

```python
df2
```

```python
df2 = df2.rename({'Amount':'y'}, axis = 'columns')
```

```python
df2
```

```python
# changed the data set so it only goes to end of 2020- wanting to see different results from a smaller sample. 
df3=df2.iloc[:18]
```

```python
df3
```

```python
m = Prophet(seasonality_mode='additive').fit(df3)
future = m.make_future_dataframe(periods=365)
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
m = Prophet(seasonality_mode='multiplicative', mcmc_samples=365).fit(df3)
fcst = m.predict(future)
fig = m.plot_components(fcst)
```

```python
#Running the new model with the main data file.
m = Prophet(seasonality_mode='additive').fit(df2)
future = m.make_future_dataframe(periods=3652)
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
m = Prophet(seasonality_mode='additive', mcmc_samples=300).fit(df2)
fcst = m.predict(future)
fig = m.plot_components(fcst)
```

```python
future = m.make_future_dataframe(periods=120, freq='MS')
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
df_actual = df3
m = Prophet(seasonality_mode='additive', mcmc_samples=365).fit(df3)
future = m.make_future_dataframe(periods=365)
fcst = m.predict(future)
```

```python
se = fcst.loc[0:17, 'yhat'] - df_actual['y']
mse = np.mean(se)
rmse = np.sqrt(mse)
```
