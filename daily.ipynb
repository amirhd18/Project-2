```python
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
#Load date in from CSV file

df_2019 = pd.read_csv("../Project_2/2019.csv")
df_2020 = pd.read_csv("../Project_2/2020.csv")
df_2021 = pd.read_csv("../Project_2/2021.csv")
```

```python
#show data is reading correctly for each file

df_2019.head()
```

```python
df_2020.head()
```

```python
df_2021.head()
```

```python
#combined all data into one data set

df = pd.concat([df_2019,df_2020,df_2021])
```

```python
df.columns
```

```python
#chage the column names to "ds" and "y" for the following steps using Prophet

df = df.rename({"Date":"ds","Amount":"y"}, axis='columns')
```

```python
df.columns
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
m = Prophet(seasonality_mode='multiplicative').fit(df)
future = m.make_future_dataframe(periods=3652)
fcst = m.predict(future)
fig = m.plot(fcst)
```

```python
m = Prophet(seasonality_mode='multiplicative', mcmc_samples=300).fit(df)
fcst = m.predict(future)
fig = m.plot_components(fcst)
```
